```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import savgol_filter

# Define a function with a hump, an increasing trend, and noise
def noisy_hump_trend(x):
    hump = np.exp(-(x - 3)**2 / 2)  # Gaussian hump centered at x=3
    trend = 0.5 * x  # Linear increasing trend
    noise = 0.3 * np.random.rand(len(x)) 
    return hump + trend + noise

# Generate data points
x = np.linspace(0.0, 5.0, 100)
y = noisy_hump_trend(x)

# Apply Savitzky-Golay filter
window_length = 21  # Adjust window size as needed
polynomial_order = 2  # Adjust polynomial order as needed
filtered_y = savgol_filter(y, window_length, polynomial_order)

# Plot original and filtered data
plt.plot(x, y, label='Noisy Data')
plt.plot(x, filtered_y, label='Filtered Data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Filtering Noise from a Function with a Hump and Trend')
plt.grid(True)
plt.show()
```

```python
# Define a function with two humps, decreasing trend, sharp hump, and increasing trend
def complex_function(x):
    hump1 = 0.8 * np.exp(-(x - 2)**2)  
    hump2 = 0.6 * np.exp(-(x - 3.5)**2 / 0.5)  
    decreasing_trend = -0.4 * (x - 4)  
    sharp_hump =  np.exp(-(x - 6)**2 / 0.2)  
    increasing_trend = x - 5
    noise = 0.2 * np.random.randn(len(x))
    return hump1 + hump2 + decreasing_trend + sharp_hump + increasing_trend + noise

# Generate data points
x = np.linspace(0, 8, 200)  # Increased range for wider function
y = complex_function(x)

# Apply Savitzky-Golay filter 
window_length = 20  # Potentially adjust window size
polynomial_order = 3  # Potentially adjust polynomial order
filtered_y = savgol_filter(y, window_length, polynomial_order)

# Plot the results
plt.plot(x, y, label='Complex Function with Noise')
plt.plot(x, filtered_y, label='Filtered Data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Filtering a Complex Function with Savitzky-Golay')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
import matplotlib.ticker as ticker
from scipy.signal import butter, filtfilt

test_number= 23
# Function to apply a simple moving average filter
def moving_average(signal, window_size):
    window = np.ones(int(window_size)) / float(window_size)
    return np.convolve(signal, window, 'same')

def butter_lowpass_filter(data, cutoff, fs, order=5):
    nyq = 0.5 * fs  # Nyquist Frequency
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    y = filtfilt(b, a, data)
    return y

# Assuming 'df' has been prepared and loaded with the 't' and 'derp' columns
time_columns = [col for col in df.columns if col.startswith('t')]
derivative_columns = [col for col in df.columns if col.startswith('derp')]

plt.figure(figsize=(14, 8))

# Define noise levels
noise_levels = [0.2, 0.4, 0.6, 0.7]
test_number = 20  # Example test number, adjust as needed
for i, noise_level in enumerate(noise_levels):
    # Extract the time and derivative data corresponding to the test number
    time_data = df[time_columns[test_number]]
    derivative_data = df[derivative_columns[test_number]]
    
    # Add noise to the derivative data
    noise = np.random.normal(0.2, noise_level, size=derivative_data.shape)
    noisy_signal = derivative_data + noise
    
    # Apply Savitzky-Golay filter
    window_length = 40 if len(noisy_signal) > 20 else len(noisy_signal) // 2 * 2 + 2
    polynomial_order = 9
    filtered_signal = savgol_filter(noisy_signal, window_length, polynomial_order)
    
    # Apply moving average filter to the Savitzky-Golay filtered signal
    #smoothed_signal = moving_average(filtered_signal, window_size=40)  # Adjust window size as needed

    # Apply Butterworth low-pass filter
    fs = 70  # Sample rate, or the number of measurements per unit of time
    cutoff = 3.5  # Desired cutoff frequency of the filter, Hz
    smoothed_signal = butter_lowpass_filter(noisy_signal, cutoff, fs, order=2)

    # Plotting
    ax = plt.subplot(2, 2, i + 1)
    ax.loglog(time_data, derivative_data, label='Original Signal', color='blue')
    ax.loglog(time_data, noisy_signal, label='Noisy Signal', color='orange', alpha=0.6)
    ax.loglog(time_data, filtered_signal, label='Filtered Signal', color='green', alpha=0.5)
    ax.loglog(time_data, smoothed_signal, label='Smoothed Signal', color='red', alpha=0.8)
    ax.set_title(f'Noise level: {noise_level}')
    ax.legend()

    # Add minor ticks and grid
    ax.xaxis.set_minor_locator(ticker.LogLocator(subs='auto'))
    ax.yaxis.set_minor_locator(ticker.LogLocator(subs='auto'))
    ax.grid(True, which="minor", ls="--")
    
    # Set axis labels for all plots
    ax.set_xlabel('Time', fontsize=12)
    ax.set_ylabel('Derivative', fontsize=12)

plt.tight_layout()
plt.show()
```
